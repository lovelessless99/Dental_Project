```python
def new_source(path_list, mode):
        new_source = lambda x : x.split("/")[2].replace(f'_{mode}',"") + '_' + '_'.join(x.split("/")[-1].split("_")[1:]) 
        path_list = [ new_source(i) for i in path_list]
        return path_list
```

```python
directory = [ 
                "Dental_Data/PBL/10_interdental_20200901_Max_4_stage"           ,  
                "Dental_Data/PBL/10_interdental_20200901_Max_4_stage_Flip"      , 
                "Dental_Data/PBL/10_interdental_clahe_20200901_Max_4_stage"     , 
                "Dental_Data/PBL/10_interdental_clahe_20200901_Max_4_stage_Flip"
            ]


HV_directory = [
                "Dental_Data/PBL/10_interdental_20200901_Max_4_HV"           ,  
                "Dental_Data/PBL/10_interdental_20200901_Max_4_HV_Flip"      , 
                "Dental_Data/PBL/10_interdental_clahe_20200901_Max_4_HV"     , 
                "Dental_Data/PBL/10_interdental_clahe_20200901_Max_4_HV_Flip"
            ]


directory = [ i + "/mapping.json" for i in directory]
HV_directory = [ i + "/mapping.json" for i in HV_directory]

HV_data = load_json(HV_directory, interdental=True)
HV_dataset = json_2_dataframe_PBL_inderdental(HV_data)

data = load_json(directory, interdental=True)
dataset = json_2_dataframe_PBL_inderdental(data)

dataset["source"] = new_source(dataset["Path"], "stage")
HV_dataset["source"] = new_source(HV_dataset["Path"], "HV")

result = pd.merge(dataset, HV_dataset, how='inner', on=['source', 'Class', 'tooth_num', 'angle'])
result.rename(columns={'Path_x': 'Path', 'State_x': 'State', 'State_y': 'bone_loss'}, inplace=True)
result = result.drop(columns="Path_y")
result = result[["Path", "tooth_num", "State", "Class", "bone_loss"]]
```
