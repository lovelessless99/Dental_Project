```python

model_info = ("Inception", Inception)
performance_list = []

Fold_num = 5
classes = 2
num_params = 1
fold_data_num = 1

for train_dataset, valid_dataset, test_dataset, train_generator, valid_generator, test_generator in\
        prepared_data(classes, Fold_num, batch_size=64):
        
        valid_dataset = valid_dataset[~valid_dataset.ID.isin(Sheng_ID)]
        test_dataset = test_dataset[~test_dataset.ID.isin(Sheng_ID)]
        
        valid_generator = make_generator(valid_dataset, batch_size=64)
        test_generator  = make_generator(test_dataset, batch_size=64)
            
        K.clear_session()
        best_models = None
        
        model_name, model_fnc = model_info
        K.clear_session()

        param_dir  = f"{model_name}_Fold_{fold_data_num}"
        print(f"Fold {fold_data_num}: loading {model_name} ......")

        start = time.time()
        model = model_fnc(input_shape=(200, 180, 1), classes=len(np.unique(train_dataset["Class"])))
        end = time.time()
        elapse = end - start
        print(f"Fold {fold_data_num}: loading Done, cost {elapse} seconds")


        K.clear_session()
        top_k_param = get_k_top_value(f"Results/{root_dir}/Fold_{fold_data_num}/{param_dir}/parameter",  k_th = num_params )
        best_param = top_k_param[0]
        
        K.clear_session()
        print(f"loading {best_param} ......")
        start = time.time()
        best_model = load_model(best_param)
        end = time.time()
        elapse = end - start
        print(f"loading Done, cost {elapse} seconds")

        valid_predict = best_model.predict_generator(valid_generator, verbose=1)
        valid_predict = valid_predict[:len(valid_dataset)]
        valid_result  = np.argmax(valid_predict, axis=1)
        
        valid_dataset["Predict"] = valid_result
        
        K.clear_session()
        
        confusion = confusion_matrix(valid_dataset.Class.astype('int'), valid_result.astype('int'))
        curr_acc = (valid_dataset.Class == valid_dataset.Predict).mean() * 100
        title = "Accuracy  = {:5.2f} %".format(curr_acc)
        print(title)
        init_directory(f"Results/{root_dir}/Fold_{fold_data_num}/table")
        plot_confusion_matrix(confusion, len(np.unique(valid_dataset.Class)), curr_acc, f"Results/{root_dir}/Fold_{fold_data_num}", 0)
        
        valid_predict = np.round(valid_predict, 2)
        
#         test_dataset[f"{model_name}_Prob"] = pd.Series(test_predict.tolist())
#         valid_dataset[['Class_0_prob', 'Class_1_prob', 'Class_2_prob']] = pd.DataFrame(valid_predict)
        valid_dataset[['Class_0_prob', 'Class_1_prob']] = pd.DataFrame(valid_predict)
        valid_dataset["Predict"] = valid_result
        
        valid_dataset.to_csv(f"Results/{root_dir}/Fold_{fold_data_num}/table/valid_predict_no_Sheng.csv", index=False)
        detail = statistic(valid_dataset)
        detail.to_csv(f"Results/{root_dir}/Fold_{fold_data_num}/table/val_cmx_stage_no_Sheng.csv")
        print(detail)
        
        
        K.clear_session()
        
        performance_list.append(curr_acc)
        fold_data_num += 1
```
