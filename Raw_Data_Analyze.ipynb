```python
def load_json_normal(normal_path, interdental=False):
            mapping_data = json.load(open(normal_path, "r"))
            filter_fun = lambda x : { path: max(list(map(int, state))) for path, state in x.items() if max(list(map(int, state))) >= 0 }
            interdental_fun = lambda x : { path: state for path, state in x.items() if state >= 0 }
            filter_normal_data       = filter_fun(mapping_data) if not interdental else interdental_fun(mapping_data)
            filter_data = collections.OrderedDict()

            for normal in filter_normal_data.keys():
                    filter_data[normal]      = filter_normal_data[normal]
            return filter_data
```

```python
directory = "Dental_Data/PBL/10_20200901_Original/mapping.json"

data = load_json_normal(directory)
dataset = json_2_dataframe_PBL(data)
result = statisatic_dental_for_class(dataset)
result.T.astype(int)
```

```python
in_directory = "Dental_Data/PBL/10_interdental_20200901_Max_4_Original/mapping.json"

in_data = load_json_normal(in_directory, interdental=True)
in_dataset = json_2_dataframe_PBL_inderdental(in_data)
in_result = statisatic_dental_for_class(in_dataset)
in_result.T
```
